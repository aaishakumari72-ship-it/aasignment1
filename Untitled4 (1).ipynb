Q1. Explain the difference between Relative , Absolute and Mixed Cell Referencing.

>>Relative - The default type. It changes when you copy the formula.Notation is  A1,If a formula in B1 contains =A1 and is copied to B2, the new formula becomes =A2.

>>Absolute - A fixed reference. It does not change when you copy the formula.
Notation is $ A$1,If a formula in B1 contains =$A$1 and is copied to B2, the formula in B2 will still be =$A$1.

>>Mixed - A hybrid type where either the column or the row is locked, but not both.Notation is  $A1 OR A$1,If a formula in B1 contains=A$1  and  is copied to B2, the new formula will be =A$1 . If the formula contains $A1 and is copied to B2, the new formula will be $A2 .
